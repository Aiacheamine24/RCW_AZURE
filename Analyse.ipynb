```python
# External Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Internal Imports

# Read in the data
dataset = pd.read_csv('./private/files/heart.csv')
data = dataset.copy()
```

## Analyse Exploratoire

```python
# Display the first 5 rows of the dataset
print(data.head())
# Display the last 5 rows of the dataset
print(data.tail())
# Display the shape of the dataset
print(data.shape)
# Display the columns of the dataset
print(data.columns)
# Display the info of the dataset
print(data.info())
```

```python
## Descriptive Statistics
# Display the summary statistics of the dataset
print(data.describe(include='all'))
```

```python
# Count NaN values in the dataset
data.isna().sum()
```

 # Data Visualization

```python
# Calcul des moyennes par rapport à la cible
mean_data = data.groupby('target').mean()

# Affichage du graphique
mean_data.reset_index().plot(x='target', y=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal'], kind='bar', figsize=(9, 5))
plt.title("Mean Positive Heart Disease Depending age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal")
plt.xlabel("Targets")
plt.ylabel("Mean")
plt.show()
```

```python

# Calcul des moyennes par rapport à la cible
mean_data = data.mean()

# Tri des moyennes dans l'ordre décroissant
mean_data_sorted = mean_data.sort_values(ascending=False)

# Affichage des données triées
print(mean_data_sorted)

# Tracer un diagramme circulaire (pie chart) des moyennes
mean_data_sorted.plot.pie(autopct='%1.1f%%', title='Repartition des moyennes des variables')
plt.show()
```
